```python
#left rectangle
from math import *
import  matplotlib.lines
import pylab
import numpy as np
def function(x):
    return sin(x*x)
def drawLine (axes,x,y):
    line  = matplotlib.lines.Line2D (x, y, color="k")
    axes.add_line (line)
def searchIntegralLeftRectangle(a,b,n,axes):
    h = (b-a)/n
    x = [a+i*h for i in range(n+1)]
    I = 0
    for i in range(n):
        I+=function(x[i])*(x[i+1] - x[i])
        drawLine(axes,[x[i] , x[i]] , [0,function(x[i])])
        drawLine(axes,[x[i+1] , x[i+1]] , [0,function(x[i])])
        drawLine(axes,[x[i] , x[i+1]] , [function(x[i]) , function(x[i])])
    return I
a = 0
b = pi/4
x = np.linspace(a,b,20)
pylab.plot(x,[function(i) for i in x],'g')
pylab.xlim(0,1)
pylab.ylim(0,0.75)
pylab.grid()
axes = pylab.gca()
I = searchIntegralLeftRectangle(a,b,20,axes)
pylab.show()
print('I = ' ,I )
```

```python
#right triangle
def searchIntegralRightRectangle(a,b,n,axes):
    h = (b-a)/n
    x = [a+i*h for i in range(n+1)]
    I = 0
    for i in range(1,n+1):
        I+=function(x[i])*(x[i] - x[i-1])
        drawLine(axes,[x[i-1] , x[i-1]] , [0,function(x[i])])
        drawLine(axes,[x[i] , x[i]] , [0,function(x[i])])
        drawLine(axes,[x[i-1] , x[i]] , [function(x[i]) , function(x[i])])
    return I
a = 0
b = pi/4
x = np.linspace(a,b,20)
pylab.plot(x,[function(i) for i in x],'b')
pylab.xlim(0,1)
pylab.ylim(0,0.75)
pylab.grid()
axes = pylab.gca()
I = searchIntegralRightRectangle(a,b,20,axes)
pylab.show()
print('I = ' ,I )
```

```python
#middle rectangle
def searchIntegralMiddleRectangle(a,b,n,axes):
    h = (b-a)/n
    x = [a+i*h for i in range(n+1)]
    I = 0
    for i in range(n):
        I+=function((x[i]+x[i+1])/2)*(x[i+1] - x[i])
        drawLine(axes,[x[i] , x[i]] , [0,function((x[i]+x[i+1])/2)])
        drawLine(axes,[x[i+1] , x[i+1]] , [0,function((x[i]+x[i+1])/2)])
        drawLine(axes,[x[i] , x[i+1]] , [function((x[i]+x[i+1])/2) , function((x[i]+x[i+1])/2)])
    return I
a = 0
b = pi/4
x = np.linspace(a,b,20)
pylab.plot(x,[function(i) for i in x],'r')
pylab.xlim(0,1)
pylab.ylim(0,0.75)
pylab.grid()
axes = pylab.gca()
I = searchIntegralMiddleRectangle(a,b,20,axes)
pylab.show()
print('I = ' ,I )


```
